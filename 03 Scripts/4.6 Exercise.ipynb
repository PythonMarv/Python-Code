```python
# Import combined Dataset
df_ords_combined = pd.read_pickle(os.path.join(path,'02 Data','Prepared Data','orders_products_combined.pkl'))
```

```python
# Import products Dataset
df_prod = pd.read_csv(os.path.join(path,'02 Data','Prepared Data','products_checked.csv'), index_col = False)
```

## Step 5. Merging df_prods with df_ords_combined

```python
# Drop unnamed columns in df_prods
df_prod = df_prod.drop(columns = 'Unnamed: 0')
```

```python
# Drop unnamed columns in df_ords_combined
df_ords_combined = df_ords_combined.drop(columns = ['Unnamed: 0','_merge'])
```

```python
# Merge df_prod and df_ords_combined
df_merged = df_prod.merge(df_ords_combined, on = 'product_id', indicator = True)
```
